### 1.데이터 프레임 생성

```python
#data 딕셔너리의 키가 데이터프레임의 열 인덱스가 됨
data={
    '국어':[80,90,70,30],
    '영어':[90,70,60,40],
    '수학':[90,60,80,70]
}
df=pd.DataFrame(data)
df
```

```python
df.index=['춘향','몽룡','향단','방자']
df
```
