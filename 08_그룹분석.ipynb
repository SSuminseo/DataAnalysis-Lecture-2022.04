# 데이터 프레임-그룹 분석

```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
iris=sns.load_dataset('iris')
tips=sns.load_dataset('tips')
```

# 1.Groupby

# Iris 데이터 사례

```python
iris.tail()
```

```python
#각 품종별 feature의 평균
iris.groupby('species').mean()
```

```python
iris.groupby('species').first() #각종의 첫번째 
```

```python
#각 품종별 sepla_length의 표준편차
iris.groupby('species').std()['sepal_length']
```

```python
iris.groupby('species').std()[['sepal_length']]
```

```python
iris[['sepal_length','species']].groupby('species').std()
```

```python
iris.groupby('species')[['sepal_length']].std()
```

-그룹 연산을 여러가지 하는 경우
   -각 품종별 sepal_length의 평균,표준편차,최소값,최대값

```python
iris.groupby('species')['sepal_length'].agg(['mean','std','min','max'])
```

# 2. Pivot table

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
df1.pivot_table('인구','도시','연도')
```

```python
df1.pivot_table('인구','도시','연도',margins=True,margins_name='평균')
```

```python
df1.pivot_table('인구',index=['연도','도시'])
```

-tips 데이터를 통해 groupby,pivot_table 사용법 비교

```python
# 성별과 흡연유무에 따른 데이터의 갯수
tips.groupby(['sex','smoker']).size()
```

```python
tips.pivot_table('tip_pct','sex','smoker',aggfunc='count',margins=True,margins_name='계')
```

```python
#성별에 따른 평균 팁 비율
tips.groupby('sex')[['tip_pct']].mean()
```

```python
tips.pivot_table('tip_pct', 'sex') #알아서 평균 구해줌-->기본 연산이 평균임
```

```python
#흡연유무에 따른 평균 팁 비율과 최소,최대값
tips.groupby('smoker')[['tip_pct']].agg(['mean','min','max'])
```

```python
tips.pivot_table('tip_pct','smoker',aggfunc=['mean','min','max'])
```

```python
#성별과 흡연유무에 따른 평균 팁 비율
tips.pivot_table('tip_pct',['sex','smoker'])
```

```python
tips.pivot_table('tip_pct','sex','smoker')
```

```python
ages=[0,2,10,21,23,37,61,20,41,32,100]
bins=[1,20,30,50,70,100]
labels=['미성년자','청년','중년','장년','노년']
cats=pd.cut(ages, bins, labels=labels)
cats
```

```python
print(cats.categories)
print(cats.codes)
```

```python
adf=pd.DataFrame(ages, columns=['ages'])
adf['age_cat']=pd.cut(adf.ages, bins,labels=labels)
adf
```
